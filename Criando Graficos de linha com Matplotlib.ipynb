# Importanto bibliotecas e criando sequencia de valores

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
valores = np.linspace(start=0,stop=5,num=10,endpoint=False,retstep=True)
#num = número de elementos gerandos na linha
#endpoint = false não retorna o número 5 (stop point)
#redstep = true retorna o intervalo entre os numeros neste caso foi 0,5
print(valores)
print('')
#abaixo retornamos somente os dados
print(valores[0])
```

### Agora que temos um conjunto de números de 0 a 4.5 vamos criar um gráfico rapidamente

Utilizando-se da funcao plot da biblioteca plt passamos como parametro os números obtidos anteriormente e temos um gráfico!

```python
eixo_y = valores[0]
plt.plot(eixo_y)
```

podemos adicionar legenda e mostrar o grafico novamente

```python
plt.plot(eixo_y)
plt.ylabel("Numeros gerados anteriormente")
plt.xlabel("quantidade de numeros na sequencia")
```

Observe que temos os valores passados de 0 a 4.5 mas também é possível ver que o matplotlib gera automaticamente uma lista com o mesmo número de valores (neste caso 10 valores) no eixo x para plotar o grafico, assumindo que é uma sequencia de valores a partir do 0 (zero)

## Quando desejarmos podemos passar os valores de x e y para que o gráfico exiba o desejado.

```python
import numpy as np
eixo_x = np.linspace(start=10,stop=100,num=10,endpoint=True)
```

```python
eixo_x
```

```python
plt.plot(eixo_x,eixo_y)
plt.ylabel("Numeros gerados anteriormente")
plt.xlabel("quantidade de numeros na sequencia")
```

Para cada conjunto de argumentos x e y existe um terceiro que é a formatação que pode ser indicada pela cor e/ou tipo de plotagem. <br> Por exemplo: o parametro "r^" definirá aqui a formatacao de cor vermelha (red) e um triângulo como tipo de plotagem.

```python
plt.plot(eixo_y,eixo_x,'r^')
plt.ylabel("Numeros gerados anteriormente")
plt.xlabel("quantidade de numeros na sequencia")
```

também é possível deixar os valores dos pontos independentes dos eixos passando os parametros dos pontos ordenados em plot e definindo somento os eixos com o parametro axis e informando os valores máximos de cada eixo.

```python
plt.plot(eixo_y,pow(eixo_y,2),'r^')
plt.axis([-1, 10, -3, 30]) # neste caso -1 a 10 para x e -3 a 30 para y
print(eixo_y)
print(pow(eixo_y,2))
```

para acessar outros tipos de formatacao consulte: http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

Entretanto, observe que o ponto de partida fica solto no espaço, por isso é interessante adicionarmos um grid para melhor visualização. Palavras chave usadas como parametros/argumentos (kwargs) podem ser adicionadas ao grid. Para encontrar outras kwargs consulte: kwargs control the Line2D properties em http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot/

```python
plt.plot(valores[0],pow(valores[0],2),'r^')
plt.axis([-1, 10, -3, 30]) # neste caso -1 a 10 para x e -3 a 30 para y
plt.grid(animated=True,alpha=0.4,color='b',linestyle='solid')
```

## Gráfico de multiplas linhas / sequências de dados

Usando np.arange() para criar uma sequencia uma nova sequencia de dados

```python
linha = np.arange(0.,3.,0.1)
```

```python
linha
```

### Criando gráfico de multiplas linhas representando funcção linear, quadrada e cúbica

```python
linear=linha
quadratica = linha**2
cubica = linha**3
plt.plot(linha,linear,'r--',linha,quadratica,'bs',linha,cubica,'g^')
```

aqui estamos usando o ipython para imprimir o grafico na tela mas se quisermos fazer isso manualmente devemos usar a função plt.show()

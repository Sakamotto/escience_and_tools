```python
import warnings
import seaborn as sns
import pandas as pd
%matplotlib inline
warnings.simplefilter(action="ignore",category=UserWarning)
warnings.simplefilter(action="ignore",category=Warning)
```

```python
titanic = sns.load_dataset("titanic")
titanic1 = pd.read_csv("titanic.csv")
titanic2= pd.read_csv("titanic2.csv")
```

```python
sns.heatmap(iris.corr(),cmap='coolwarm')
```

```python
sns.heatmap(iris.corr(),cmap='coolwarm',annot=True)
```

```python
sns.pairplot(iris)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.pairplot(tips,hue='sex',palette='husl')
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot,n_levels=6)
```

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris);
```

```python
sns.jointplot(y='pclass',x ='age',data=titanic,kind="hex", color="k")
```

```python
sns.lmplot(x='petal_length', y='petal_width', data=iris);
```

```python
from scipy import stats
g = sns.JointGrid(x="petal_length", y="petal_width", data=iris)
g = g.plot(sns.regplot, sns.distplot)
g = g.annotate(stats.pearsonr)
```

```python
sns.lvplot(x='petal_length', y='petal_width', data=iris);
```

```python
sns.lvplot(x='sepal_length', y='sepal_width', data=iris);
```

```python
stats.spearmanr(tips.tip,tips.total_bill)
```

```python
stats.pearsonr(tips.tip,tips.total_bill)
```

```python
coeficiente_pearson = stats.pearsonr(tips.tip,tips.total_bill)[0]
coeficiente_spearman = stats.spearmanr(tips.tip,tips.total_bill)[0]
```

```python
import matplotlib.pyplot as plt
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot_joint(plt.scatter, color="g", s=40, edgecolor="white")
g = g.plot_marginals(sns.distplot, kde=False, color="g")
g = g.annotate(stats.pearsonr)
```

```python
from scipy import stats
import matplotlib.pyplot as plt
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot_joint(plt.scatter, color="g", s=40, edgecolor="white")
g = g.plot_marginals(sns.distplot, kde=False, color="g")
g = g.plot(sns.regplot, sns.distplot)
g = g.annotate(stats.spearmanr)
```

```python
ax = sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
ax = sns.regplot(x="total_bill", y="tip", data=tips,ci=coeficiente_pearson)
```

```python
ax = sns.regplot(x="total_bill", y="tip", data=tips,ci=coeficiente_spearman)
```
